## Histograms

```python
import matplotlib.pyplot as plt
import numpy as np

data=np.random.random_sample(5)
plt.hist(data,color="orange",edgecolor="red",bins=5) #use plt.hst to create histogram
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram Plot Example")

plt.show()
```

## ScatterPlot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,15,13,17,20]
plt.scatter(x,y,marker="o",color='blue') #use plt.scatter to create scatter plot
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot Example")
plt.show()
```

## PieChart

```python
categories=['Housing','Transportation','Food','Entertainment','Utilities']
expenses=[1200,300,400,200,150]

plt.figure(figsize=(5,16))
plt.pie(expenses,labels=categories,autopct="%2.2f%%") #use plt.pie for creating pie chart
plt.title("Monthly Expenses By Category")
plt.legend()
plt.show()
```

## BoxPlot 

#### Boxplot for summarize the distribution of numerical data by displaying quartiles,outliers, and potential skewness

```python
import matplotlib.pyplot as plt

data=[12,15,10,15,17,18,20,22,23,25,27,30,35,100,-34]
plt.boxplot(data) #createing boxplot
plt.title("Boxplot Example with Sample Data")
plt.xlabel("Data")
plt.ylabel("Values")
plt.show()
```

## Stack Plot

- Also Known as Stacked area plot.
- its a type od data visualization that displays multiple datasets as layers stacked on top of another

```python
import matplotlib.pyplot as plt
quarters=['Q1','Q2','Q3','Q4']
electronics=[10000,12000,11000,10500]
clothing=[5000,6000,7500,8000]
home_appliances=[3000,7500,8200,9000]

plt.figure(figsize=(10,6))
plt.stackplot(quarters,electronics,clothing,home_appliances,labels=['Electronics','Clothing','Home Appliances'])
plt.xlabel("Quarters")
plt.ylabel("Sales ($)")
plt.title("Product Category Sales Over Quarters")
plt.legend()
plt.show()
```

 ## 3D Plot

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)
x,y=np.meshgrid(x,y)
z=np.sin(np.sqrt(x**2 + y**2))
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(x,y,z,c=z,cmap="viridis")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
plt.show()
```

```python
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
y3=np.tan(x)
y4=np.exp(-x)

#create a 2x2 subplot
fig,axs=plt.subplots(2,2,figsize=(10,8))
axs[0,0].plot(x,y1,'r') #top-left subplot
axs[0,0].set_title("Sine Wave")

axs[0,1].plot(x,y2,'g') # top-right subplot
axs[0,1].set_title("Cosine Wave")

axs[1,0].plot(x,y3,'y') #bottom -left subplot
axs[1,0].set_title("Tangent Wave")

axs[1,1].plot(x,y4,'m') # bottom - right subplot
axs[1,1].set_title("Exponential Decay")
plt.tight_layout()
plt.show()
```
