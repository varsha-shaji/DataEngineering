```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Line Plot

months=['Jan','Feb','Mar','Apr','May','Jun']
traffic=[150,2,180,220,250,310]

plt.plot(months,traffic) #creating line plot
plt.show()
```

```python
# Adding labels and title

months=['Jan','Feb','Mar','Apr','May','Jun']
traffic=[150,2,180,220,250,310]

plt.plot(months,traffic) #creating line plot


plt.xlabel("Month")
plt.ylabel("Monthly Traffic")
plt.title("Monthly Website Traffic")
plt.show() #displays the plot

```

```python
# Adding styles to graph


months=['Jan','Feb','Mar','Apr','May','Jun']
traffic=[150,2,180,220,250,310]

plt.plot(months,traffic,marker="^",markersize=10,color="blue",linestyle="-",linewidth=3) # adding styles


plt.xlabel("Month")
plt.ylabel("Monthly Traffic")
plt.title("Monthly Website Traffic")
plt.show() #displays the plot
```

```python
months=['Jan','Feb','Mar','Apr','May','Jun']
product_a_revenue=[45,55,40,70,80,75]
product_b_revenue=[35,40,50,55,70,68]

plt.plot(months,product_a_revenue,marker="o",linestyle="-",color="orange",label="Product A")
plt.plot(months,product_b_revenue,marker="s",linestyle="--",color="red",label="Product B")

plt.xlabel("Month")
plt.ylabel("Monthly Revenue (in $1000)")
plt.title("Monthly Revenue Comparison")
plt.legend()  #display a legend to differentiate b/w product A and Product B
plt.show()  
```

```python
people=['Amiya','Tom','sara','Ann','Jacob','Judin']
bp_values1=[67,85,80,78,69,67]
bp_values2=[122,124,130,127,118,125]
plt.plot(people,bp_values1,marker="o",markersize=5,linestyle="-",color="orange",label="Diastolic Pressure")

plt.plot(people,bp_values2,marker="d",markersize=5,linestyle="--",color="purple",label="Systolic Pressure")
plt.title("People Blood Pressure Comparison")
plt.legend()
plt.savefig("BP_LinePlot.pdf") #without specifying extension it will be png
plt.show()
```

```python
# Plot Types

#Bar plot
import matplotlib.pyplot as plt
categories=['Category A','category B','Category C','Category D']
values=[20,35,30,25]

plt.bar(categories,values,color=['red','g','black'],width=0.8)
plt.xlabel("Categories")
plt.ylabel('Values')
plt.title("Bar Plot Example")

plt.show()
```

```python
# Plot Types

#Bar plot Horizontal use barh instead of bar
import matplotlib.pyplot as plt
categories=['Category A','category B','Category C','Category D']
values=[20,35,30,25]

plt.barh(categories,values,color=['red','g','black'])
plt.xlabel("Categories")
plt.ylabel('Values')
plt.title("Bar Plot Example")

plt.show()
```

```python
import pandas as pd
df=pd.read_csv(r"C:/Users/Administrator/Desktop/python/IPL_Matches_2008-2020.csv")
df
```

```python
import matplotlib.pyplot as plt
matches_played=pd.concat([df['team1'],df['team2']]).value_counts()
wins=df['winner'].value_counts()

plt.bar(matches_played.index, matches_played.values, color="red", label="Matches Played Count")
plt.bar(wins.index, wins.values, color="blue", label="Win Count")

plt.xlabel("Team")
plt.ylabel("Count")
plt.title("IPL Matches Played vs Wins by Team (2008-2020)")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt


player_counts = df['player_of_match'].value_counts()

plt.figure(figsize=(12,6))
plt.bar(player_counts.index[:10], player_counts.values[:10], color="red", label="Player of Match Count")

plt.xlabel("Player")
plt.ylabel("Count")
plt.title("Top 10 IPL Player of the Match Winners (2008-2020)")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
