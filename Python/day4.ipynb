```python
def name_length(name):
    if len(name)>=8 and len(name)<=16:
        return True
    else:
        return False


def name_str(name):
    if isinstance(name,str):
        return True
    else:
        return False

def first_char_validation(name):
    if (name[0]).isupper():
        return True
    else:
        return False
```

```python
import unittest
class UserNameValidation(unittest.TestCase):
    def test_name_length_positive(self):
        self.assertTrue(name_length("abcdefgd"))
        print("Test Passed:Expected length for username")
    def test_name_length_negative1(self):
        self.assertFalse(name_length("abcfgd"))
        print("Test Passed:username doesnt have atleast 8 character")
    def test_name_length_negative2(self):
        self.assertFalse(name_length("djejriamruehfsajfhasd"))
        print("Test Passed:username have more than 16 characters")

    def test_name_str1(self):
        self.assertTrue(name_str("Abcdefgs"))
        print("Test Passed:username is string")
    def test_name_str_negative1(self):
        self.assertFalse(name_str({}))
        print("Test Passed:username must be string")

    def test_first_char_positive(self):
        self.assertTrue(name_str('Ahdfbksdfof'))
        print("Test Passed:username starts with capital letter")

    def test_first_char_negative1(self):
        self.assertTrue(name_str('1dfbksdfof'))
        print("Test Passed:username must start with capital letter")
if __name__ == "__main__": 
    unittest.main(argv=[''],verbosity=2, exit=False)
        
```

```python
class MyList:
    def __init__(self):
        self.lst = [1, 2, 3, 4,5,6]
        self.index = 0  

    def __iter__(self):
        return self 

    def __next__(self):
        if self.index < len(self.lst):
            value = self.lst[self.index] ** 2  
            self.index += 1
            return value
        else:
            raise StopIteration 

cls = MyList()
for x in cls:
    print(x)
```

```python
Generatorspecial typeof iterator that is defined using  functionand yield keyword

used yield instead of return
generates values lazily(one at a time)
memory efficient than normal functions
```

```python


def my_generator():
    for i in range(1,6):
        yield i

gen=my_generator()
for value in gen:
    print(value)
```

```python
def my_generator():
    for i in range(1,10):
        if i%2==0:
            yield i


gen=my_generator()
for value in gen:
    print(value)
```

iterator vs generator(difference)

iterator:requires class with __iter__() and __next__()
generator:uses function with yield keyword
iterator:mpre code ,manual implementation
generator:simple and concise implementation
iterator:slightly more control
generator: more memory efficient (memory optimized)

```python
class MyList:
    def __init__(self):
        self.lst=[2,3,4,5,6,7]
        self.index=0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index< len(self.lst):
            val=self.lst[self.index]*5
            self.index+=1
            return val
        else:
            raise StopIteration

cls=MyList()
it=iter(cls)
for i in it:
    print(i)
```
