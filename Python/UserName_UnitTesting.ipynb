```python
def name_length(name):
    return len(name)>=8 and len(name)<=16
        
def name_str(name):
   return isinstance(name,str)

def first_char_validation(name):
    return (name[0]).isupper()
```

```python
import unittest
class UserNameValidation(unittest.TestCase):
    def test_name_length_positive(self):
        self.assertTrue(name_length("abcdefgd"))
        print("Test Passed:Username have required length")
    def test_name_length_negative1(self):
        self.assertFalse(name_length("abcfgd"))
        print("Test Passed:username doesn't have atleast 8 character")
    def test_name_length_negative2(self):
        self.assertFalse(name_length("djejriamruehfsajfhasd"))
        print("Test Passed:username have more than 16 characters")

    def test_name_str1(self):
        self.assertTrue(name_str("Abcdefgs"))
        print("Test Passed:username is string")
    def test_name_str_negative1(self):
        self.assertFalse(name_str(12))
        print("Test Passed:username must be string")

    def test_first_char_positive(self):
        self.assertTrue(first_char_validation('Ahdfbksdfof'))
        print("Test Passed:username starts with capital letter")

    def test_first_char_negative1(self):
        self.assertFalse(first_char_validation('dfbksdfof'))
        print("Test Passed:username must start with capital letter")
if __name__ == "__main__": 
    unittest.main(argv=[''],verbosity=2, exit=False)
        
```
