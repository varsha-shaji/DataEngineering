```python

def get_hcf(a,b):
    while b!=0:
        a,b=b,a%b
    return a

print(get_hcf(2,3))
    
```

```python
def factorial(n):
    fact=1
    for i in range(2,n+1):
        fact*=i
    return fact

print(factorial(5))
```

```python
# isinstance() 
#map()
# map applies a function to all items in input list
# syntax:map(function,list_of_input)

# case-1
def hallsquare(num):
    return (num**2)

squared=list(map(hallsquare,[1,2,3,4,5]))
print(squared)

# case-2

squared=list(map(lambda x:x**2,[1,2,3,4,5]))
print(squared)
```

```python
# reduce()
# is for performing some computation on a list and returning the list

product=1
lst=[1,2,3,4]
#case-1
for num in lst:
    product*=num
print(product)

#case-2
from functools import reduce
def multiply(x,y):
    return x*y;
product=reduce(multiply,lst)
print(product)


#case-3:
product=reduce(lambda x,y:x*y,lst)  # it work like accumulator
print(product)
```

```python
# Keyword Argument

# kwargs -allow to pass keywords variable length of arguments to function.
# *kwargs-single argument
# **kwargs-multiple arguments

def lst(**kwargs):
    if kwargs:
        print("Contact {0},{1},{2},{3}".format(kwargs['name'],kwargs['mob'],kwargs['add'],kwargs['course']))


lst(name="prwatech",mob=7808365790,add="Banglore_Pune",course="Technical")
```

```python
# Arbitrary Arguments
# Allows to pass n no of arguments to function

def lst(*names):
    print(names)
    for lst2 in names:
        print("Name {0}".format(lst2))

lst("same","hari","john",1)
```

```python
class Sports:
    def __init__(self,model,color,cost):
        self.__model=model
        self.__color=color
        self.__cost=cost

    def getModel(self):
        return self.__model

    def setModel(self,model):
        self.__model=model

    def getColor(self):
        return self.__color
    def setColor(self,color):
        self.__color=color

    def getCost(self):
        return self.__cost

    def setCost(self,cost):
        self.__cost=cost

class Shoes(Sports):
    def __init__(self,model,color,cost,brand):
        super().__init__(model,color,cost)
        self.__brand=brand
    def ShoesInfo(self):
        return self.getModel()+" of brand "+self.getCost()+" in "+self.getColor()+" color for "+self.getCost()

    sh1=Shoes("Shoes","Black","Rs.3699","Xyz123")
    print(sh1.ShoesInfo())
    print(sh1.getModel())
    sh2=Shoes("Shoes","Black-red","Rs.4599","Xyz345")
    print(sh2.ShoesInfo())
```

```python
import sys
try:
    a=int(input("Enter first Number :"))
    b=int(input("Enter Second Number:"))
    ch=input("Enter a operation (+,-,*,/): ")
    if ch=="+":
        res=a+b
    
    elif ch=="-":
        res=a-b
        
    elif ch=="*":
        res=a*b
    elif ch=="/":
        try:
            res=a/b
            
        except:
            print(sys.exc_info())
            sys.exit(1)
    else:
        print("invalid operation")
    print("{0} {1} {2} is {3}".format(a,ch,b,res))

except:
    print(sys.exc_info())
```
