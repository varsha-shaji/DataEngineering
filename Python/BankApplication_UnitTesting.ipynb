```python
class Bank:
    def __init__(self,balance=0):
        if not isinstance(balance,(int,float)):
            raise ValueError("only integer or float value is allowed")
        elif balance<0:
            raise ValueError("Cannot initialize with amount less than zero")
            
        self.balance=float(balance)
    def deposit(self,amt):
        if not isinstance(amt,(int,float)):
            raise ValueError("only integer or float value is allowed")
        elif amt<=0:
            raise ValueError("Amount must be greater than Zero")
        self.balance+=amt

    def withdraw(self,amt):
        if amt<=0:
            raise ValueError("Amount must be greater than zero")
        elif amt>self.balance:
            raise ValueError("Insufficient Balance")
        self.balance-=amt

    def get_balance(self):
        return float(self.balance)
```

```python
import unittest
class TestBankApplication(unittest.TestCase):
    def test_get_balance(self):
        acc=Bank(100)
        bal=acc.get_balance()
        self.assertEqual(bal,100)
        print(bal)
        

    def test_deposit_positive(self):
        acc=Bank(100)
        dep=acc.deposit(100)
        self.assertEqual(acc.balance,200)
        
    def test_deposit_zero(self):
        acc = Bank(100)
        with self.assertRaises(ValueError):
            acc.deposit(0)
       
 
    def test_deposit_negative(self):
        acc = Bank(100)
        with self.assertRaises(ValueError):
            acc.deposit(-10)
 
    def test_withdraw_valid(self):
        acc = Bank(100)
        acc.withdraw(40)
        self.assertEqual(acc.balance, 60)
 
    def test_withdraw_zero(self):
        acc = Bank(100)
        with self.assertRaises(ValueError):
            acc.withdraw(0)
 
    def test_withdraw_more(self):
        acc = Bank(50)
        with self.assertRaises(ValueError):
            acc.withdraw(100)
 
    def test_withdraw_negative(self):
        acc = Bank(100)
        with self.assertRaises(ValueError):
            acc.withdraw(-20)

    

unittest.main(argv=[''],verbosity=2, exit=False)
```
