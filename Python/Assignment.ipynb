```python
Assignment-1-Transpose of a matrix with and without using  built-in function


#with function
import numpy as np
matrix=[
    [1,2,3,4],
    [6,8,3,4],
    [7,8,9,3]
]
transposed=np.transpose(matrix)
print(transposed)



#without function
matrix=[
    [1,2,3,4],
    [6,8,3,4],
    [7,8,9,3]
]
transposed=[[matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix[0]))]
print(transposed)



```

Assignment 1-why we cannot call the local variable from outside the function


we cant use local variable outside the function because of scope
local variable is a varaibale that exists only inside that function and it get destroyed after the function finishes.

inorder to use it outside  make the local varibale as global using global keyword or

```python
# without global
def test():
    x=10

print(x) #error
```

```python
#using global
def test():
    global x
    x=20
test()
print(x)
```

Assignment-2 where do use map,filter and reduce

```python
# map()
# use map() when to apply a function to each element
# and the output list will have the same no .of elements

nums=[1,2,3,4]
squared=list(map(lambda x:x*x ,nums))
print(squared)

#real time use case-Calculating tax for a list of prices.
```

```python
# filter()
# used when you want to select some items that satisfy a condition.output will have fewer elements

nums=[1,2,3,4,5,6]
evens=list(filter(lambda x: x%2==0,nums))
print(evens)

#real time use case-Keeping only valid email addresses.
```

```python
# reduce()
# used when yoy want to combine all items into one value.accumulating values.

from functools import reduce
nums=[1,2,3,4]
sum=reduce(lambda x,y:x+y,nums)
print(sum)

#real time use case-: Determining the highest score in a game.
```
