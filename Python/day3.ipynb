```python
f=open('test.txt') 
print(f.read(4)) #read from start to 8th character
f.tell()  #tell() will tell the position of pointer ,point to the last position of substring
f.close()
```

```python
f=open('test.txt') 
print(f.read(8)) #read from start to 8th character
print(f.tell()) 
f.seek(5) #seek set the position to the one specified
f.tell()
f.close()
```

```python
f=open('test.txt') 
print(f.readline()) #read first line
print(f.readlines()) # reads all lines
f.close()
```

```python
#renaming and deleting files
#rename
import os
os.rename("test.txt","sample.txt")
f=open("sample.txt","r")
print(f.readlines())
f.close()
```
