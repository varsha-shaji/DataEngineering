```python
lst={1,2,3,4,5,6,9}
lst2={6,7,8,9,0}
print(lst | lst2) #either | or union function
print(lst.union(lst2))

print(lst&lst2) #either intersection function or &,it will give common elements from both
print(lst.intersection(lst2))

print(lst-lst2)#either difference function or ,it will give common elements from both
print(lst.difference(lst2))

print(lst ^ lst2)
print(lst.symmetric_difference(lst2))#either intersection function or &,it will give not common elements from both
```

```python
#Assignment-1-Transpose of a matrix with and without using  built-in function

#with function
import numpy as np
matrix=[
    [1,2,3,4],
    [6,8,3,4],
    [7,8,9,3]
]
transposed=np.transpose(matrix)
print(transposed)
```

```python
#without function
matrix=[
    [1,2,3,4],
    [6,8,3,4],
    [7,8,9,3]
]
transposed=[[matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix[0]))]
print(transposed)
```
